```python
reponse = requests.get(url)
```

```python
titres = []
liens = []
prix = []
en_stock = []
notes = []
soup = BeautifulSoup(reponse.text)
articles = soup.findAll('article')
for article in articles:
        titre = article.find('img').get('alt')
        lien = article.find('img').get('src')
        price = article.find('p', class_='price_color').get_text(strip=True)
        stock = article.find('p', class_='availability').get_text(strip=True)
        rating = ' '.join(article.find('p', class_='star-rating')\
            .get('class')).replace('star-rating', '').strip()
        titres.append(titre)
        liens.append('https://books.toscrape.com/' + lien)
        prix.append(price)
        en_stock.append(stock)
        notes.append(rating)

print(titres)
```

```python
books = []
books.append(titres)
books.append(prix)
books.append(notes)
books.append(en_stock)
books.append(liens)


print(books)
```
